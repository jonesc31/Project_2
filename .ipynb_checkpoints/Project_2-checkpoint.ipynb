```python
# Set the ticker
ticker = "SPY"

# Set timeframe to "1Day"
timeframe = "1Day"

# Set start and end datetimes of 3 years
start_date = pd.Timestamp("2021-01-02", tz="America/New_York").isoformat()
end_date = pd.Timestamp("2024-01-02", tz="America/New_York").isoformat()

# Get 5 years worth of historical data for SPY
ticker_data = api.get_bars(
    ticker,
    timeframe,
    start=start_date,
    end=end_date
).df
ticker_data.head()
```

```python
# Filter the date index and close columns

spy_df = ticker_data.loc[:, ["open", "high", "low", "close","volume" ]]

# Use the pct_change function to generate returns from close prices
spy_df["Actual Returns"] = spy_df["close"].pct_change()

# Drop all NaN values from the DataFrame
spy_df = spy_df.dropna()

# Review the DataFrame
display(spy_df.head())
display(spy_df.tail())
```

```python
# Plot the DataFrame with hvplot
spy_df["close"].hvplot()
```

```python
#Recreate Trading Algo 

# Create a signals_df DataFrame

signals_df = spy_df.copy()

# Set the short window and long windows
short_window = 20
long_window = 50

# Add the HMA technical indicators for the short and long windows
signals_df["Short"] = TA.HMA(signals_df, short_window)
signals_df["Long"] = TA.HMA(signals_df, long_window)

# Review the DataFrame
signals_df.iloc[95:105, :]
```

```python
# Set the Signal column
signals_df["Signal"] = 0.0

# Generate the trading signal 1 or 0,
# where 1 is when the Short window is greater than (or crosses over) the Long Window
# where 0 is when the Short window is under the Long window
signals_df["Signal"][short_window:] = np.where(
    signals_df["Short"][short_window:] > signals_df["Long"][short_window:], 1.0, 0.0
)

# Calculate the points in time at which a position should be taken, 1 or -1
signals_df["Entry/Exit"] = signals_df["Signal"].diff()

# Review the DataFrame
signals_df.iloc[95:105, :]
```

```python
# Visualize entry position relative to close price
entry = signals_df[signals_df["Entry/Exit"] == 1.0]["close"].hvplot.scatter(
    color='purple',
    marker='^',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize exit position relative to close price
exit = signals_df[signals_df["Entry/Exit"] == -1.0]["close"].hvplot.scatter(
    color='orange',
    marker='v',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize close price for the investment
security_close = signals_df[["close"]].hvplot(
    line_color='lightgray',
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize moving averages
moving_avgs = signals_df[["Short", "Long"]].hvplot(
    ylabel='Price in $',
    width=1000,
    height=400
)

# Overlay plots
entry_exit_plot = security_close * moving_avgs * entry * exit
entry_exit_plot
```

```python
### Create a New Stratrgy Using Keltner Indicator From Finta Library
```

```python
# Create a new clean copy of the signals_df DataFrame
kc_signals_df = spy_df.copy()

# Review the DataFrame
kc_signals_df.head()
```

```python
# Determine the Keltner Bands for the Dataset
kc_df = TA.KC(kc_signals_df)

# Review the DataFrame
kc_df.iloc[17:25, :]
```

```python
# Concatenate the Keltner Bands to the DataFrame
kc_signals_df = pd.concat([kc_signals_df, kc_df], axis=1)

# Review the DataFrame
kc_signals_df.iloc[17:25, :]
```

```python
# Visualize close price for the investment
security_close = kc_signals_df[["close"]].hvplot(
    line_color='lightgray',
    ylabel='Price in $',
    width=1000,
    height=400
)

kc_upper = kc_signals_df[["KC_UPPER"]].hvplot(
    line_color='red',
    ylabel='Price in $',
    width=1000,
    height=400
)

kc_lower = kc_signals_df[["KC_LOWER"]].hvplot(
    line_color='green',
    ylabel='Price in $',
    width=1000,
    height=400
)


# Overlay plots
kcbands_plot = security_close * kc_upper * kc_lower
kcbands_plot
```

```python
# Update the trading algorithm using Keltner Channels 'KC'

# Set the Signal column
kc_signals_df["Signal"] = 0.0

# Create a value to hold the initial trade signal
trade_signal = 0

# Update the DataFrame Signal column 1 (entry) or -1 (exit) for a long position trading algorithm
# where 1 is when the Close price is less than the KC_LOWER window
# where -1 is when the Close price is greater the the KC_UPPER window
# Incorporate a conditional in the if-statement, to evaluate the value of the trade_signal so the algorithm 
# plots only 1 entry and exit point per cycle.
for index, row in kc_signals_df.iterrows():
    if (row["close"] < row["KC_LOWER"]) and (trade_signal < 1):
        kc_signals_df.loc[index, "Signal"] = 1.0
        trade_signal += 1
        
    if (row["close"] > row["KC_UPPER"]) and (trade_signal > 0):
        kc_signals_df.loc[index, "Signal"] = -1.0
        trade_signal = 0


# Review the DataFrame
kc_signals_df
```

```python
# Visualize entry position relative to close price
entry = kc_signals_df[kc_signals_df["Signal"] == 1.0]["close"].hvplot.scatter(
    color='green',
    marker='^',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize exit position relative to close price
exit = kc_signals_df[kc_signals_df["Signal"] == -1.0]["close"].hvplot.scatter(
    color='red',
    marker='v',
    size=200,
    legend=False,
    ylabel='Price in $',
    width=1000,
    height=400
)

# Visualize close price for the investment
security_close = kc_signals_df[["close"]].hvplot(
    line_color='lightgray',
    ylabel='Price in $',
    width=1000,
    height=400
)

kc_upper = kc_signals_df[["KC_UPPER"]].hvplot(
    line_color='purple',
    ylabel='Price in $',
    width=1000,
    height=400
)

kc_lower = kc_signals_df[["KC_LOWER"]].hvplot(
    line_color='blue',
    ylabel='Price in $',
    width=1000,
    height=400
)


# Overlay plots
kcbands_plot = security_close * kc_upper * kc_lower * entry * exit
kcbands_plot
```

```python
# Update the trading algorithm using Bollinger Bands

# Set the Signal column
kc_signals_df["Signal"] = 0.0

# Create a value to hold the initial trade signal
trade_signal = 0

# Update the DataFrame Signal column 1 (entry) or -1 (exit) for a long position trading algorithm
# where 1 is when the Close price is less than the BB_LOWER window
# where -1 is when the Close price is greater the the BB_UPPER window
# Incorporate a conditional in the if-statement, to evaluate the value of the trade_signal so the algorithm 
# plots only 1 entry and exit point per cycle.
for index, row in kc_signals_df.iterrows():
    if (row["close"] < row["KC_LOWER"]) and (trade_signal < 1):
        kc_signals_df.loc[index, "Signal"] = 1.0
        trade_signal += 1
        
    if (row["close"] > row["KC_UPPER"]) and (trade_signal > 0):
        kc_signals_df.loc[index, "Signal"] = -1.0
        trade_signal = 0


# Review the DataFrame
kc_signals_df
```
